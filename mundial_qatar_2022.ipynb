## Analisis

### Primera Parte

```python
equipo : str = "ARGENTINA"
```

```python
# Seleccionar todos los partidos que jugo el equipo de "local"
# Completar

filtroLocal : pd.Series = mundial["team1"] == equipo
```

```python
# Seleccionar todos los partidos que jugo el equipo de "visitante"
# Completar

filtroVisitante : pd.Series = mundial["team2"] == equipo
```

```python
# Seleccionar todos los partidos que jugo el equipo (de "local" o "visitante")
# Completar

partidos : pd.DataFrame = mundial[filtroLocal | filtroVisitante]
partidos
```

```python
# Cuantos goles hizo el equipo en total (obtener los goles metidos de local y sumarlos con los goles metidos de visitante)
# Completar


partidos_local: pd.DataFrame = mundial[filtroLocal]
partidos_visitante: pd.DataFrame = mundial[filtroVisitante]


goles_local: int = partidos_local["number of goals team1"].sum()

goles_visitante: int = partidos_visitante["number of goals team2"].sum()

goles = goles_local + goles_visitante

goles
```

```python
# Cuantos goles hizo el equipo en promedio
# Completar

cantPartidosLocal = len(partidos_local)

cantPartidosVisitante = len(partidos_visitante)

cantPartidosTotal : int = cantPartidosLocal + cantPartidosVisitante

promedioGoles : int = goles / cantPartidosTotal

promedioGoles
```

```python
# Cuantos partidos gano el equipo
# Completar

#este fue mi primer codigo y me tiraba error, entonces le pedi a chatGpt que me lo corrija.
# ganados : int
# if partidos_local["number of goals team1"] > partidos_local["number of goals team2"] :
#     ganados += 1
# elif partidos_visitante["number of goals team1"] > partidos_visitante["number of goals team2"] :
#     ganados += 1
# ganados

ganados = 0

ganados_local : int= (partidos_local["number of goals team1"] > partidos_local["number of goals team2"]).sum()

ganados_visitante : int = (partidos_visitante["number of goals team1"] < partidos_visitante["number of goals team2"]).sum()

ganados : int = ganados_local + ganados_visitante

ganados
```

```python
# Le pedi a chat que me ayude a hacer el grafico

# Graficar los goles hechos por el equipo en cada partido
# Obtener solo los goles hechos por Argentina en cada partido
goles_argentina = partidos["number of goals team1"].where(filtroLocal, partidos["number of goals team2"])
partidos["goles_argentina"] = goles_argentina
partidos = partidos.dropna()

# Ordenar por fecha
partidos = partidos.sort_values(by="date")

# Filtrar las fechas donde se marcaron goles de Argentina
fechas_con_goles = partidos[partidos["goles_argentina"] > 0]["date"]

# Graficar los goles hechos por Argentina en cada fecha donde se marcaron goles
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(partidos["date"], partidos["goles_argentina"], marker='o', linestyle='-', color='lightgray')
ax.scatter(fechas_con_goles, partidos.loc[partidos["goles_argentina"] > 0, "goles_argentina"], color='blue', label="Goles de Argentina")
ax.set_xlabel("Fecha")
ax.set_ylabel("Goles")
ax.set_title(f"Goles hechos por {equipo} en cada partido")
plt.xticks(fechas_con_goles, rotation=45)
plt.tight_layout()
plt.legend()
plt.show()

# Completar
```

### Segunda Parte

```python
# Contar cuantos goles hizo cada equipo en total en el mundial
# Completar

# En este ejercicio me inspire en chatGPT para hacer el ultimo for

equipos = set(mundial["team1"]).union(mundial["team2"])

# Diccionario para almacenar los goles de cada equipo y calcular los goles totales
goles = {}

for equipo in equipos :

    goles_totales = goles_local + goles_visitante
     
    goles[equipo] = goles_totales

for equipo, total_goles in goles.items():
    print(f"{equipo}: {total_goles} goles")
```

```python
# Graficar los goles por equipo en todo el mundial
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(goles.keys(), goles.values())
ax.set_xlabel("Equipo")
ax.set_ylabel("Total de Goles")
ax.set_title("Goles por Equipo en el Mundial")
plt.xticks(rotation=45, ha='right')  # Rotar las etiquetas del eje x para mayor legibilidad
plt.tight_layout()
plt.show()

# Completar
```
