```python
# Seleccionar todos los partidos que jugo el equipo de "visitante"
# Completar

filtroVisitante : pd.Series = mundial["team2"] == equipo
```

### Segunda Parte

```python
# Contar cuantos goles hizo cada equipo en total en el mundial
# Completar

# En este ejercicio me inspire en chatGPT para hacer el ultimo for y en la idea del diccionario
equipos = set(mundial["team1"]).union(mundial["team2"])

# Diccionario para almacenar los goles de cada equipo y calcular los goles totales
goles = {}

for equipo in equipos:

    filtroLocal = mundial["team1"] == equipo
    filtroVisitante = mundial["team2"] == equipo
    partidos_local = mundial[filtroLocal]
    partidos_visitante = mundial[filtroVisitante]
    
    # Suma de los goles del equipo como local y visitante
    goles_local = partidos_local["number of goals team1"].sum()
    goles_visitante = partidos_visitante["number of goals team2"].sum()
    
    # Goles totales del equipo
    goles_totales = goles_local + goles_visitante
    
    # Poner los goles en el diccionario
    goles[equipo] = goles_totales

# Printear total de goles de cada equipo
for equipo, total_goles in goles.items():
    print(f"{equipo}: {total_goles} goles")

```

```python
# Graficar los goles por equipo en todo el mundial

fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(goles.keys(), goles.values())
ax.set_xlabel("Equipo")
ax.set_ylabel("Total de Goles")
ax.set_title("Goles por Equipo en el Mundial")
plt.xticks(rotation=45, ha='right')  # Rotar las etiquetas del eje x para mayor legibilidad
ax.set_ylim(0, 20)  # Establecer el rango del eje y
plt.tight_layout()
plt.show()

# Completar
```

```python
# Cual es el equipo que hizo mas goles en el mundial?
# Completar

# Me inspire en chatGPT

equipoMasGoles = max(goles, key=goles.get)
cantGoles = goles[equipoMasGoles]

print(f"El equipo que hizo más goles en el mundial fue {equipoMasGoles} con un total de {cantGoles} goles.")
```
